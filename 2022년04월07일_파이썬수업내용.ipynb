````python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np
# 폰트 지정
font_location = "C:\\Windows\\Fonts\\malgun.ttf"
fontname = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family = fontname)
tips = sns.load_dataset("tips") # 제공 받는 데이터 예제
tips # 실행문

sns.countplot(x="day", data=tips)
plt.title("팁스 요일별 값")
plt.show()
````

````python
# 요일별 식사 금액 (박스플롯을 이용하여 표시)
sns.boxplot(x="day",y="total_bill", data=tips)
plt.show()
````

````python
# 흡연자와 비흡연자 비교, 요일별 식사 금액
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips)
plt.title("흡연자 비흡연자")
plt.show()
````

````python
# 남자와 여자 비교, 요일별 식사 금액
sns.boxplot(x="day", y="total_bill", hue="sex", data=tips)
plt.title("남자, 여자의 요일별 식사 금액")
plt.show()
````

````python
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips)
plt.show()
````

````python
sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
plt.show()
````

````python
sns.set_style("dark")
sns.lmplot(x="total_bill", y="tip", data=tips, size=5)
plt.show()
````

````python
# 식사 금액과 팁의 흡연자 비흡연자 기준, 색 변경
sns.set_style("dark")
sns.lmplot(x="total_bill", y="tip", hue="smoker" ,data=tips, size=5, palette="Set1")
plt.show()
````

````python
# 히트맵 사용
flights = sns.load_dataset("flights") # 제공 받는 데이터 예제
print(flights.head(10))
print(type(flights))
````

````python
section_data = flights.pivot("month", "year", "passengers") # index, colune, value 순으로 지정함.
section_data
````

````python
# 데이터를 가지고 시각화(히트맵 표시)
plt.figure(figsize=(10,10))
sns.heatmap(section_data, annot=True, fmt='d', cmap="RdYlGn_r") # 숫자 표시, 정수로 표시 
# YlGnBu 파란색과 노란색 계열
# RdYlGn_r
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
tips = sns.load_dataset("tips")
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True) # jitter = True 무작위로 겹치지 않게 출력
plt.show()
````

````python
# 데이터 변수가 여러개인 다차원 데이터는 데이터 종류에 따라 다름
# 1. 분석하고자 하는 데이터가 모두 실수인 경우
# 2. 분석하고자 하는 데이터가 모두 카테고리 값인 경우
# 3. 분석하고자 하는 데이터가 모두 실수값과 카테고리 값이 섞여있는 경우
# jointplot
import seaborn as sns
# sns.jointplot(x="xname" , y="yname" , data = dataframe, kind = "scatter")
sns.jointplot(x="day", y="total_bill", data=tips_data, kind="scatter")
plt.show()
````
