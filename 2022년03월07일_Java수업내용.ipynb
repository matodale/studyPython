### 객체 선언에서의 this
```java
public class exam03 {

	public static void main(String[] args) {
		MyCustom ch1 = new MyCustom(); // 객체 선언
		MyCustom ch2 = new MyCustom();
		MyCustom ch3 = ch1;
		MyCustom ch4 = ch2;
		ch3.age = 20;
		ch3.isMarried = false;
		ch3.name = "김철수";
		ch4.age = 21;
		ch4.isMarried = false;
		ch4.name = "박희영";
		
		Person ch5 = new Person(); // 객체 선언
		ch5.age = 23;
		ch5.isMarried = false;
		ch5.name = "이기철";
		ch5.introduce(); // 객체 함수 호출
		ch5.name = "박철호";
		ch5.introduce(); // 객체 함수 호출
	}
}
class MyCustom { // 클래스 선언
	int age;
	boolean isMarried;
	String name;
	
}
class Person { // 클래스 선언
	int age;
	boolean isMarried;
	String name;
	void introduce() { // 메서드 method 선언
		int age = this.age; 
		boolean isMarried = this.isMarried;
		System.out.println("안녕하세요. 저는 50살 미혼 철수 입니다.");
		System.out.println("=== 자기 소개 ===");
		System.out.printf("이름 : %s\n",this.name); // 해당 클래스 변수를 가져옴.
		System.out.printf("나이 : %d\n",age);
		System.out.printf("결혼 여부 : %b\n",isMarried);
	}
}
```

### 객체란
* 실체 : 실제 존재하는 것
* 주체 : 실체가 자신을 부를 때.
* 객체 : 실체가 다른 실체를 바래볼 때.

```java
class A { // 상위 클래스 선언
	int x, y;
	int fn(int x, int y) {
		while (x <= 10) {
			System.out.println(x + "번째");
			while (y <= 3) {
				System.out.println(y);
				y++;
			}
			y = 1;
			x++;
		}
		return 0;
	}
}
class B extends A{ // 상속
	int x1, y1;
	
	int fn(int x, int y) {
		x = this.x1;
		y = this.y1;
		System.out.println("for문으로 쓰일때");
		for(x= 1; x<10; x++) {
			System.out.println(x + "번째");
			for(y=1; y<4; y++) {
				System.out.println(y);
			}
			y = 1;
		}
		return 0;
	}
}

public class exam01 {
	public static void main(String[] args) {
		// 문제 : 1부터 3까지 출력하는 작업을 10번 해주세요.
		// 조건 : 2중 while문 사용
		int i=1,j=1; // 초기값 변수 선언
		A a1 = new B(); // for문으로 실행과 while 문으로 실행문 두가지
		a1.fn(i, j); 
	}
}
```
