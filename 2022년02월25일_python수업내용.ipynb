```python
# while 버전
h = 3 # 전체 높이 및 넓이
l = 1 # 해당 라인에 그려야 할 별의 개수
# 한줄에 별의 개수만큼 별을 그리는 반복문
while l <= h: #  높이만큼 반복
    i = 0 # 그린 별이 개수를 세기 위한 변수
    while i < l : # 해당 라인에 그려야 할 별의 개수만큼 반복
        print("*", end="") # 한줄에 그려야 하기 때문에 줄바꿈 X
        i = i + 1 # 그린 별의 개수 증가
    print() # 한줄이 완성되면 다음줄로 줄바꿈
    l = l + 1 # 별의 개수는 하나씩 증가
```

```python
# while 버전

# 전체 높이 및 넓이
h = 3 

# 해당 라인에 그려야 할 별의 개수
star_len = 1

# 높이 및 넓이만큼 반복
while star_len <= h:
    
    # 해당 라인에 그려야 할 공백의 갯수
    space_len = h - star_len 
  
    # 한줄에 공백을 그리는 반복문
    i = 0
    
    # 해당 라인에 그려야 할 공백의 갯수만큼 반복
    while i < space_len : 
        print(" ", end="") # 가로로 이어져서 그려져야 하기 때문에 줄바꿈 X
        i = i + 1 # 그린 공백의 갯수

    # 한줄에 별을 그리는 반복문
    j = 0
    while j < star_len : # 해당 라인에 그려야 할 별의 개수만큼 반복
        print("*", end="") # 가로로 이어져서 그려져야 하기 때문에 줄바꿈 X
        j = j + 1 # 그린 별의 개수 증가
  
    print() # 한줄이 완성되면 다음줄로 바꿔야하므로 줄바꿈 O

    star_len = star_len + 1 # 별이 개수는 줄이 바뀔 때마다 하나씩 증가
```

```python
height = 10
empty = 1 
for empty in range(empty, height): # empty 공백은 1 부터 10까지 증가
    space = height - empty 
    for i in range(0, space): # i 는 해당 공간만큼씩 공백을 만듬.
        print(" ", end="")
    for j in range(0, empty): # j 는 empty 만큼 *를 만듬. 
        print("*", end="")
    print()
```
