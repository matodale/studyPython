```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np
# 폰트 지정
font_location = "C:\\Windows\\Fonts\\malgun.ttf"
fontname = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family = fontname)

# 막대 그래프
plt.style.use('ggplot')
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
x = [3, 4]
y = [1, 2]
# x축 항목 이름 지정
labels = ['english', 'math']
# 컬러 지정
colors = ['green' , 'red']
# bar() 막대그래프를 그리는 함수
# x축, y축, 각 막대의 라벨, 각 막대의 컬러
ax1.bar(x, y, tick_label= labels, color = colors)
for i, j in enumerate(y):
    ax1.text(j+2, i+1.05, str(j), color = 'red', fontweight = 'bold')
plt.show()
```

```python
# 막대 그래프
plt.style.use('ggplot')
fig2 = plt.figure()
ax2 = fig2.add_subplot(111)
labels = ['english' , ' math', 'langauge']
# 컬러 지정
colors = ['green' , 'red', 'blue']
x = [3 , 4 , 5]
y = [1 , 2, 4]
ax2.barh(x,y, tick_label = labels, color = colors)
plt.show()
```

```python
import numpy as np
x = [1, 2]
y1, y2, y3 = [1,2], [2,5], [3, 7]
fig = plt.figure()
ax = fig.add_subplot(111)
w = 0.2
ax.bar(x,y1,width=w, label = 'english')
ax.bar(np.array(x)+w, y2, width=w, label = 'math')
ax.bar(np.array(x)+w*2, y3, width=w, label = 'langauge')
ax.legend() # 왼쪽 상단의 색상에 따른 라벨표
plt.show()
```

```python
# pie 차트(원 그래프) 만들기

plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
label = ['english', 'math', 'langauge']
sizes = [45, 35, 20] # 각 숫자의 합이 100
ax.pie(sizes, labels = label, autopct='%1.1f%%')
plt.show()
```

```python
# pie 차트(원 그래프) 만들기

plt.style.use('ggplot')
fig = plt.figure(constrained_layout = True)
ax = fig.add_subplot(111)
label = ['english', 'math', 'langauge']
sizes = [45, 35, 20] # 각 숫자의 합이 100
# 원차트의 함수 설정
ax.pie(sizes, labels = label, autopct='%1.1f%%', shadow=True)
# 차트 제목 설정
ax.set_title('나의 공부 성향')
plt.show()
```

```python
# box plot 그리기
plt.style.use('ggplot')
fig = plt.figure(constrained_layout = True)
ax = fig.add_subplot(111)
x = [1, 2, 3, 3, 11 , 20]
ax.boxplot(x)
plt.show()
```

```python
# 여러 개의 box plot 그리기
plt.style.use('ggplot')
fig = plt.figure(constrained_layout = True)
ax = fig.add_subplot(111)
x = [[1, 2, 3, 3, 11 , 20],[1, 2, 8, 19, 13]]
label = ['english' , 'langauge']
ax.boxplot(x, labels= label)
plt.show()
```
