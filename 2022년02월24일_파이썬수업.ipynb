## while문 복습하기

초기값  
while 조건식 :  
    참일 경우 반복될 문장  
    증감표현  


```python
i = 1 # 초기값 표현
while(i <= 5): # 반복문 변수 i 는 5까지 조건문에 의해 실행됨.
    print(i) # 출력 
    i+=1
```

```python
# 8단 구구단, 3단 구구단 표현 출력 하시오
dan_8 = 8
dan_3 = 3
i = 1
print("== 8 단 ==")
while(i<=9):
    print("{} * {} = {}".format( dan_8 , i , dan_8*i))
    i +=1
    
i = 1 
print("\n")
print("== 3 단 ==")
while(i<=9):
    print("{} * {} = {}".format(dan_3 , i , dan_3*i)) 
    i+=1
```

```python
# 2 단과 4단을 출력한다. 단 2단은 5번째까지 출력하고, 4단은 5번째를 제외한 나머지 를 출력한다.

dan_2 = 2
dan_4 = 4
i = 1
print("== 2단 ==")
while(i<=9):
    print("{} * {} = {}".format( dan_2 , i , dan_2*i))
    i+=1
    if (i==5):
        print("{} * {} = {}".format( dan_2 , i , dan_2*i))
        break
i= 1
print("== 4단 ==")
while(i<=9):
    print("{} * {} = {}".format( dan_4 , i , dan_4*i))
    i+=1          
    if i == 5:
        i+=1
        continue
        
```

```python
# 문제 : 1 부터 100까지의 수중에서 짝수의 합을 구하시오

s = 0
i = 2
while i<=100:
    s+=i
    i+=2
print(s)        
```

```python
# 문제 : 1 부터 100까지의 수중에서 짝수의 합을 구하시오
s = 0
i = 1
while i<=100: # while 문 선언
    if i%2==0:
        s+=i # s = s + i
    i+=1 # 증감
print(s)    
        
```

```python
# 문제 : 1부터 100 까지의 수중에서 짝수이고 3의 배수가 아닌 수의 합을 출력하시오
s = 0
i = 1
while i<=100:
    if i%2==0:
        if i%3!=0:
            s +=i
    i+=1
print(s) # 1734 정답   
```

```python
# 문제 : 1부터 100 까지의 수중에서 짝수이고 3의 배수가 아닌 수의 합을 출력하시오
s = 0
i = 1
while i<=100:
    if i%2==0 and i%3!=0:
        s+=i
    i+=1
print(s) # 1734 정답   
```

```python
# 문제 : 1부터 100 까지의 수중에서 짝수이고 3의 배수가 아닌 수의 합을 출력하시오
s = 0
i = 1
while i<=100:
    if i%3 == 0:
        i+=1
        continue # 3의 배수일 경우 1을 증감
    elif i%2 != 0:
        i+=1
        continue # 2의 배수 = 짝수일 경우가 아니더라도 증감
    #  둘 다 거짓일 경우에만 더한 다라는 조건
    s+=i
    i+=1
print(s)
```

## 구구단 출력하기
---

```python
i=1 # i , j  초기값 설정
j=1
while i<=9: 
    print("== {} 단 ==".format(i))
    while j<=9:
        print("{} * {} = {}".format(i , j , i*j))
        j+=1
    if i ==9 :
        print("")
        print("구구단 출력 끝입니다.")
    else:
        print("")
    j=1    
    i+=1    
```

## 구구단 출력 3중 while

```python
i=1 # i , j, k  초기값 설정
j=1
k=1
while i<=9: 
    print("== {} 단==".format(i))
    while j<=9:
        while k <=9:
            print("{} * {} * {} = {}".format(i , j , k , i*j*k))
            k+=1
        k=1
        j+=1
    j=1        
    if i == 9 :
        print("")
        print("구구단 출력 끝입니다.")
    else:
        print("")
    i+=1    
```

## 구구단을 가로로 출력하기

```python
i=1 # i , j  초기값 설정
j=1
while i<=9: 
    while j<=9:
        print("{} * {} = {}".format(j , i , i*j), end='\t')
        j+=1
    j=1    
    i+=1    
```
